```python
# determine the simulation parameter
N = 1000        # number of time steps
k = 0.1         # spring constant
m = 1           # mass
dt = 0.1        # time step
t = np.arange(0,N*dt,dt)    # time vector
```

```python
# set initial conditions
x_0 = 0
v_0 = 1

# create arrays to store the position and velocity
x = np.zeros(N)
v = np.ones(N)

def euler(x_0, v_0, dt, N):
    x[0] = x_0
    v[0] = v_0
    for i in range(1, N):
        x[i] = x[i-1] + v[i-1] * dt
        v[i] = v[i-1] - x[i-1] * k / m * dt
    return x, v

x_leap = np.zeros(N)
v_leap = np.ones(N)
x_leap_2 = np.zeros(N)
def leapFrog(x_0, v_0, dt, N):
    x_leap[0] = x_0
    v_leap[0] = v_0
    x_leap_2[0] = x_0 + v_0 * dt / 2
    for i in range(1, N):
        x_leap_2[i] = x_leap[i-1] + v_leap[i-1] * dt / 2
        v_leap[i] = v_leap[i-1] - x_leap_2[i] * k / m * dt
        x_leap[i] = x_leap_2[i] + v_leap[i] * dt / 2
    return x_leap, v_leap

x, v = euler(x_0, v_0, dt, N)
x_leap, v_leap = leapFrog(x_0, v_0, dt, N)
```

```python
# plot the position and velocity
plt.plot(t, x, label='euler method position', color='red')
plt.title('Euler and Leapfrog method - Position')
plt.xlabel('time')
plt.ylabel('position')
# add the leap frog method
plt.plot(t, x_leap, label='leap frog method position', color='blue')

plt.legend()
plt.show()
```

# Calculate the energy

```python
# Euler
E_pot = 0.5 * k * x**2
E_pot_sum = np.sum(E_pot)

E_kin = 0.5 * m * v**2
E_kin_sum = np.sum(E_kin)

E_tot = E_pot + E_kin
E_tot_sum = np.sum(E_tot)

print('E_pot_sum = ', E_pot_sum)
print('E_kin_sum = ', E_kin_sum)
print('E_tot_sum = ', E_tot_sum)

# Leapfrog
E_pot_leap = 0.5 * k * x_leap**2
E_pot_sum_leap = np.sum(E_pot_leap)

E_kin_leap = 0.5 * m * v_leap**2
E_kin_sum_leap = np.sum(E_kin_leap)

E_tot_leap = E_pot_leap + E_kin_leap
E_tot_sum_leap = np.sum(E_tot_leap)


# plot the energy of Eueler
plt.plot(t, E_pot, label='potential energy - Euler', color='red')
plt.plot(t, E_kin, label='kinetic energy - Euler', color='blue')
plt.plot(t, E_tot, label='total energy - Euler', color='green')
plt.title('Euler method - Energy')
plt.xlabel('time')
plt.ylabel('energy')
plt.legend()
plt.show()

# plot the energy of Leapfrog
plt.plot(t, E_pot_leap, label='potential energy - Leapfrog', color='red', linestyle='dashed')
plt.plot(t, E_kin_leap, label='kinetic energy - Leapfrog', color='blue', linestyle='dashed')
plt.plot(t, E_tot_leap, label='total energy - Leapfrog', color='green', linestyle='dashed')
plt.title('Leapfrog method - Energy')
plt.xlabel('time')
plt.ylabel('energy')
plt.legend()
plt.show()
```
