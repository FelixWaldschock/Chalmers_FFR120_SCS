## The Watts-Strogatz small-world model

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','grid'])
from tqdm import trange
```

```python
# parameters
n = 10             # number of nodes
p = 0.0           # probability of connection
c = 2               # number of connections to nearest neighbours (even)

# check if c is even
if c%2 != 0:
    raise ValueError("c must be even")
```

```python
# generate adjacency matrix
def generateAdjacencyMatrix(n,p,c):
    A = np.zeros((n,n))
    for i in range(n):
        for j in range(i+1,n):
            if np.random.rand() < p:
                A[i,j] = 1
                A[j,i] = 1
        # do the nearest neighbour connections
        for b in range(c):
            A[i, (i + int(b) + 1) % n] = 1

    return A
```

```python
n_a = 20
p_a = 0
c_a = 2

n_b = 20
p_b = 0.05
c_b = 2
A_a = generateAdjacencyMatrix(n_a,p_a,c_a)
A_b = generateAdjacencyMatrix(n_b,p_b,c_b)
```

```python
# plot a circle where the n nodes lay on, and draw the connections between the nodes according to the Adjacent matrix A
def plot_ErdosRenyi(A):
    n = A.shape[0]
    dAngle = 2*np.pi/n
    r = 1
    x = r*np.cos(np.linspace(0,2*np.pi,n))
    y = r*np.sin(np.linspace(0,2*np.pi,n))
    plt.figure(figsize=(6,6))
    plt.axis('off')
    plt.plot(x,y,'o',color='blue')
    for i in range(n):
        for j in range(n):
            if A[i,j] == 1:
                plt.plot([x[i],x[j]],[y[i],y[j]],color='red')
    plt.show()


plot_ErdosRenyi(A_a)
plot_ErdosRenyi(A_b)
```
